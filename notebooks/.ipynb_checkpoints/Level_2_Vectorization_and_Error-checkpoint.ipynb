## 1. Loop-Based vs Vectorized ET Computation

### Why Computation Method Matters

Computing evapotranspiration (ET) for irrigation scheduling seems straightforward: apply a formula for each day. But **how you compute** dramatically affects:
- **Speed**: Loop-based: milliseconds per calculation; Vectorized: microseconds
- **Scalability**: Can we simulate 365 days × 50 zones × 10 scenarios fast enough?
- **Real-time decisions**: Can the system respond to farmer queries instantly?

#### Loop-Based Approach (Inefficient)

Computing ET for all days using Python iteration:

```python
def compute_et_loop(temperature, wind_speed, solar_index, humidity):
    n = len(temperature)
    et = [0.0] * n
    for i in range(n):                    # Python loop overhead for EACH element
        val = 0.12*temperature[i] + 0.35*wind_speed[i] + 2.4*solar_index[i] - 0.025*humidity[i]
        et[i] = max(0.0, val)
    return et
```

**Performance cost:**
- **Loop overhead**: Python interpreter must check loop condition, resolve variable names, and manage memory **for every iteration**
- For 30 days: 30 cycles of overhead
- For 30,000 days: 30,000 cycles of overhead
- Typical speed: ~0.5–2 milliseconds per 30 days

---

#### Vectorized Approach (Efficient)

Using NumPy arrays, all computations execute in compiled C code:

```python
def compute_et_vectorized(temperature, wind_speed, solar_index, humidity):
    et_raw = (0.12*np.asarray(temperature) + 0.35*np.asarray(wind_speed)
              + 2.4*np.asarray(solar_index) - 0.025*np.asarray(humidity))
    return np.maximum(0.0, et_raw)  # All 30 (or 30,000) calculations at once
```

**Performance advantage:**
- **No loop overhead**: NumPy's C backend executes the entire array operation in a single compiled function
- CPU cache efficiency: Large arrays stay in fast L1/L2 cache
- **Typical speed: ~0.05–0.1 milliseconds per 30 days → 5–20× faster**

---

#### Real-World Scaling Example

For HydroSense-Kenya extended to 365 days, 50 irrigation zones, and 10 alternative irrigation policies:

| Operation | Method | Calculation Count | Time | Tractability |
|-----------|--------|------------------|------|--------------|
| Single simulation | Loop | 365 × 50 = 18,250 | ~36.5 sec | Slow (can't do real-time) |
| Single simulation | Vectorized | 365 × 50 = 18,250 | ~3.65 sec | Fast (acceptable) |
| Policy comparison (10 scenarios) | Loop | 182,500 | ~365 sec (6 min) | Impractical |
| Policy comparison (10 scenarios) | Vectorized | 182,500 | ~36.5 sec | Practical for optimization |

**Conclusion:** For a real-time decision-support system serving multiple farms, vectorization is **not optional—it's essential**.

````python
# Performance scaling experiment
sizes = [30, 300, 3000, 30000, 300000, 3000000]
loop_times = []
vec_times = []

for n in sizes:                                       #to test the performance of both the loop-based and vectorized implementations of the ET computation, we will run them on increasingly larger datasets. The variable `n` represents the number of data points we will use for each test. We will create larger arrays by repeating the original data until we reach the desired size `n`. Then, we will time how long it takes to compute ET using both methods and store the results for comparison.
    reps = n // 30 + 1                                #since our original data has 30 points, we calculate how many times we need to repeat it to get at least `n` points. We add 1 to ensure that we have enough data even if `n` is not an exact multiple of 30.
    T_big = np.tile(T, reps)[:n]
    W_big = np.tile(W, reps)[:n]
    S_big = np.tile(S, reps)[:n]
    H_big = np.tile(H, reps)[:n]

    T_list = T_big.tolist() 
    W_list = W_big.tolist()
    S_list = S_big.tolist()
    H_list = H_big.tolist()

    # Time loop version
    start = time.perf_counter()
    for _ in range(3):
        _ = compute_et_loop(T_list, W_list, S_list, H_list)
    loop_t = (time.perf_counter() - start) / 3
    loop_times.append(loop_t)

    # Time vectorized version
    start = time.perf_counter()
    for _ in range(3):
        _ = compute_et_vectorized(T_big, W_big, S_big, H_big)
    vec_t = (time.perf_counter() - start) / 3
    vec_times.append(vec_t)

# Display results
print(f"{'N points':<12} {'Loop (s)':<14} {'NumPy (s)':<14} {'Speedup':<10}")
print("-" * 52)
for n, lt, vt in zip(sizes, loop_times, vec_times):
    speedup = lt / vt if vt > 0 else float('inf')
    print(f"{n:<12,} {lt:<14.6f} {vt:<14.6f} {speedup:<10.1f}x")
````

````python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.loglog(sizes, loop_times, 'o-', color=COLORS['red'], linewidth=2,
           markersize=7, label='Python loop')
ax1.loglog(sizes, vec_times, 's-', color=COLORS['blue'], linewidth=2,
           markersize=7, label='NumPy vectorized')
ax1.set_xlabel('Number of data points')
ax1.set_ylabel('Execution time [seconds]')
ax1.set_title('Computation Time: Loop vs Vectorized', fontweight='bold')
ax1.legend()
ax1.grid(True, alpha=0.3)

speedups = [lt/vt for lt, vt in zip(loop_times, vec_times)]
ax2.semilogx(sizes, speedups, '^-', color=COLORS['green'], linewidth=2, markersize=8)
ax2.set_xlabel('Number of data points')
ax2.set_ylabel('Speedup factor (×)')
ax2.set_title('Vectorization Speedup', fontweight='bold')
ax2.axhline(1, color='grey', linestyle='--', alpha=0.5)
ax2.grid(True, alpha=0.3)

fig.suptitle('Performance Analysis: Why Vectorization Matters', fontsize=14, fontweight='bold')
fig.tight_layout()
plt.show()

print(f'\n📊 At 3M data points, vectorization is ~{speedups[-1]:.0f}× faster')
````

## 2. Floating-Point Arithmetic

### 2.1 Understanding Floating-Point Representation

Modern computers represent decimal numbers using **IEEE 754 double-precision** (64-bit) format:
- **Sign bit (1)**: Positive or negative
- **Exponent (11)**: Scales the number by powers of 2
- **Mantissa (52)**: The significant digits (fraction part)

This representation cannot exactly store all decimal numbers—only about $2^{53}$ distinct values across all magnitudes. Numbers like $0.1$ and $0.2$ cannot be represented exactly in binary, leading to rounding errors that accumulate through calculations.

**Why this matters for irrigation:** Each arithmetic operation introduces a small rounding error. Over 365 days of daily moisture calculations, these errors compound. A sensor reading of 30.0% might be stored as 29.99999999999997%, and small irrigation adjustments can amplify these errors.

---

### 2.2 Catastrophic Cancellation in ET Computation

The ET formula involves **subtraction of similar-magnitude terms**:

$$ET = \max\left(0,\; 0.12T + 0.35W + 2.4S - 0.025H\right)$$

When conditions are cool, cloudy, and humid—typical of early morning—the positive terms (0.12T + 0.35W + 2.4S) and negative term (−0.025H) are nearly equal in magnitude:
- Low T (15°C) contributes: 0.12 × 15 = **1.8 mm**
- Calm wind (0.5 m/s) contributes: 0.35 × 0.5 = **0.175 mm**
- Cloudy (Solar = 0.1) contributes: 2.4 × 0.1 = **0.24 mm**
- **Subtotal: ~2.2 mm**
- High humidity (95%) subtracts: 0.025 × 95 = **2.375 mm**
- **Result: 2.2 − 2.375 = −0.175 mm → max(0, ...) = 0 mm**

When subtracting numbers that are nearly identical, **floating-point cancellation** occurs: the fewer leading digits in the result means fewer significant digits of precision. An error of ±0.001 mm in either input becomes ±0.5% of the final result—significant for small ET values.

**Practical consequence:** On cool, humid mornings, ET might oscillate between 0.00 and 0.01 mm/day due to rounding alone, not real physical changes. A farmer's decision to irrigate (or not) can flip based purely on which way the rounding goes.

---

### 2.3 Accumulated Rounding Errors Over Time

The water balance recursion:
$$S_{t+1} = S_t + R_t + I_t - ET_t - D_t$$

compounds errors if computed sequentially over 365 days. A single error of ±$10^{-16}$ per day might seem negligible, but summing 365 times can accumulate to ±$10^{-14}$ or larger—enough to trigger a binary decision (irrigate/don't irrigate) at the soil moisture threshold.

**Mitigation strategies:**
1. Use **higher-precision arithmetic** (64-bit float) rather than 32-bit
2. Rearrange formulas to avoid subtracting nearly equal numbers
3. Use **Kahan summation** if accumulating many small terms
4. Validate outputs against physical bounds (e.g., ET ≥ 0)

````python
print("=== Floating-Point Demonstrations ===")
print()
a = 0.1 + 0.2
print(f"0.1 + 0.2 = {a}")
print(f"0.1 + 0.2 == 0.3? {a == 0.3}")
print(f"Difference from 0.3: {a - 0.3:.2e}")
print()

# Catastrophic cancellation in ET formula
T_low, H_high = 15.0, 95.0
W_calm, S_cloudy = 0.5, 0.1

et_terms = [0.12 * T_low, 0.35 * W_calm, 2.4 * S_cloudy, -0.025 * H_high]
print("ET term decomposition (low-T, high-H scenario):")
for name, val in zip(['0.12*T', '0.35*W', '2.4*Solar', '-0.025*H'], et_terms):
    print(f'  {name:>12s} = {val:+.4f}')
print(f"  {'Sum':>12s} = {sum(et_terms):+.6f}")
print(f"  {'max(0,sum)':>12s} = {max(0, sum(et_terms)):.6f}")
print()

# Accumulation error
print("Accumulation error over 365 daily steps:")
s_accumulated = np.float64(30.0)
daily_increment = np.float64(0.1)
for _ in range(365):
    s_accumulated += daily_increment
s_precise = 30.0 + 365 * 0.1
print(f"  Direct computation:  {s_precise}")
print(f"  Accumulated result:  {s_accumulated}")
print(f"  Difference:          {abs(s_precise - s_accumulated):.2e}")
````

## 3. Error Propagation: Temperature Uncertainty → Irrigation Recommendations

### 3.1 How Measurement Noise Propagates Through Calculations

Field sensors introduce measurement uncertainty:
- **Temperature sensors**: ±0.5°C typical uncertainty
- **Wind speed anemometers**: ±0.1 m/s typical uncertainty  
- **Soil moisture sensors**: ±2% volumetric water content
- **Solar radiation (pyranometers)**: ±5% uncertainty

These inputs feed through the ET formula:
$$\Delta ET \approx \left|\frac{\partial ET}{\partial T}\right| \Delta T + \left|\frac{\partial ET}{\partial W}\right| \Delta W + \ldots$$

Taking partial derivatives of the ET formula:
- $\frac{\partial ET}{\partial T} = 0.12$ → **±0.5°C error produces ±0.06 mm/day ET error**
- $\frac{\partial ET}{\partial W} = 0.35$ → **±0.1 m/s wind error produces ±0.035 mm/day ET error**
- $\frac{\partial ET}{\partial S} = 2.4$ → **±5% solar error produces ±0.12 mm/day ET error**

Over a 120 m² irrigation zone:
- **ET error of ±0.12 mm/day translates to ±14.4 liters/day volume error**
- A 30-day growing period compounds to **±432 liters cumulative error**—nearly 20% of a typical daily requirement

**Key insight:** Temperature is the least sensitive input; solar radiation is the most sensitive. Cloud cover uncertainty dominates ET uncertainty.

---

### 3.2 Decision Instability at Soil Moisture Thresholds

The irrigation decision rule is typically:
$$\text{Irrigate if } S < S_{\text{threshold}} \text{ else } \text{Don't Irrigate}$$

Suppose the threshold is $S_{\text{threshold}} = 22.5\%$ and today's soil moisture is $S_{\text{obs}} = 22.5\%$ (at the boundary).

Due to measurement noise ($\pm 2\%$), the *true* moisture could be anywhere in $[20.5\%, 24.5\%]$:
- If true S is 20.5%: **Irrigate**
- If true S is 24.5%: **Don't Irrigate**

**At borderline conditions, a single noisy measurement can flip the decision 50% of the time—even though the true condition hasn't changed!**

The decision is **unstable**: near thresholds, the farmer receives conflicting guidance from day to day, leading to:
1. **Over-irrigation on some days** (false positive demand)
2. **Under-irrigation on other days** (false negative demand)  
3. **Inefficient water use** and plant stress variability

---

### 3.3 Quantifying Uncertainty via Monte Carlo Ensemble Methods

To handle uncertainty rigorously, instead of a single point forecast, generate an **ensemble of plausible scenarios**:

1. **Create ensemble members**: Run 5,000 simulations where each input is perturbed by its measurement error (drawn from normal distribution)
2. **Collect ensemble predictions**: 5,000 predicted soil moistures for tomorrow ($S_{t+1}^{(i)}$)
3. **Compute ensemble statistics**:
   - **Median (50th percentile)**: Best estimate
   - **5th and 95th percentiles**: Prediction bounds  
   - **Probability $P(S < S_{\text{threshold}})$**: Likelihood irrigation is needed

**Example result:** Instead of binary "Irrigate/Don't Irrigate," report:
- "70% probability S will drop below 22.5% threshold"
- "Soil moisture likely to be 21.2% ± 2.1% (median ± std dev)"

This allows the farmer to:
- Make **probabilistic irrigation decisions** (e.g., "irrigate if probability > 60%")
- **Adjust cautiously near thresholds** rather than making binary switches
- **Reduce inefficiency** from threshold-crossing noise

**Implementation note:** Level 5 (Simulation and Optimization) will demonstrate ensemble methods with Monte Carlo sampling to robustly handle this uncertainty and optimize irrigation timing.

````python
from simulation import compute_drainage

# Fixed conditions
base_T = 25.0
base_W = 2.0
base_S = 0.7
base_H = 65.0
s0 = 28.0
field_cap = 41.0
drain_coeff = 0.18
min_moisture = 22.0
target_moisture = 33.0

noise_levels = np.linspace(0, 2.0, 20)
n_samples = 5000

et_stds = []
irr_stds = []

rng = np.random.default_rng(42)

for sigma in noise_levels:
    T_noisy = base_T + rng.normal(0, max(sigma, 1e-10), n_samples)
    et_samples = compute_et(T_noisy, np.full(n_samples, base_W),
                            np.full(n_samples, base_S),
                            np.full(n_samples, base_H))
    drainage = compute_drainage(s0, field_cap, drain_coeff)
    irr_needed = np.maximum(0, target_moisture - s0 - 3.0 + et_samples + drainage)
    et_stds.append(np.std(et_samples))
    irr_stds.append(np.std(irr_needed))

et_stds = np.array(et_stds)
irr_stds = np.array(irr_stds)

from visualization import plot_error_propagation
fig = plot_error_propagation(noise_levels, et_stds, irr_stds)
plt.show()

print("\n📊 Interpretation:")
print("  • ET uncertainty is LINEAR in temperature noise (coefficient = 0.12)")
print("  • At ±1°C error, irrigation varies by ~±0.12 mm/day")
print("  • For a 120 m² zone, this is ±14.4 L/day — non-trivial for smallholders")
````

````python
# Decision instability at borderline moisture
n_trials = 10000
rng2 = np.random.default_rng(123)
s_borderline = 22.5

T_noisy = base_T + rng2.normal(0, 0.5, n_trials)
et_noisy = compute_et(T_noisy, np.full(n_trials, base_W),
                      np.full(n_trials, base_S), np.full(n_trials, base_H))
drainage = compute_drainage(s_borderline, field_cap, drain_coeff)
s_next_noisy = s_borderline + 3.0 - et_noisy - drainage

irrigate_decision = s_next_noisy < min_moisture
flip_rate = np.mean(irrigate_decision)

fig, ax = plt.subplots(figsize=(10, 5))
ax.hist(s_next_noisy, bins=60, color=COLORS['blue'], alpha=0.7,
        edgecolor='white', density=True)
ax.axvline(min_moisture, color=COLORS['red'], linewidth=2.5, linestyle='--',
           label=f'Stress threshold = {min_moisture}%')
ax.axvline(np.mean(s_next_noisy), color=COLORS['orange'], linewidth=2,
           label=f'Mean S(t+1) = {np.mean(s_next_noisy):.2f}%')
ax.set_xlabel('Predicted soil moisture S(t+1) [% vol.]')
ax.set_ylabel('Probability density')
ax.set_title(f'Decision Instability: ±0.5°C Noise Flips Decision {flip_rate:.0%} of the Time',
             fontweight='bold')
ax.legend()
plt.tight_layout()
plt.show()

print(f'\n⚠️  Irrigate/don\'t-irrigate decision flips in {flip_rate:.1%} of simulations')
print("   → Monte Carlo ensemble methods (Level 5) address this directly")
````
